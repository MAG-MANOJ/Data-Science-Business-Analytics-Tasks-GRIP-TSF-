Which shipping mode is more widely used?

```python
sns.countplot(x=df['Ship Mode'],order=df['Ship Mode'].value_counts().index)
plt.show()
```

Bar Graph of the Average Profit or Loss and the Sales of a Category

```python
df.groupby("Category")[[ 'Sales','Profit']].sum().plot.bar(color = ["green","yellow"],figsize= (10,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

```python
df['Ship Mode'].value_counts().plot.pie(autopct='%0.1f%%',radius=1.5)
plt.show()
```

```python
df['Category'].value_counts().plot.pie(autopct='%0.1f%%',radius=1.5)
plt.show()
```

the most sold Sub-Category

```python
df['Sub-Category'].value_counts().plot.pie(autopct='%0.1f%%', radius=1.5)
plt.show()
```
